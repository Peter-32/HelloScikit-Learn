```python
import numpy as np

from sklearn import decomposition
from sklearn import datasets

import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
y[0:100]
```

```python
pca = decomposition.PCA(n_components=2)
pca.fit(X)
pcaX = pca.transform(X)
```

```python
debuggingGetFirstColumn = pcaX[:,0]
debuggingGetFirstColumn[0:10]
```

```python
def colorAssigner(x):
    if (x==0):
        return 'b'
    elif (x==1):
        return 'g'
    else:
        return 'r'
colorAssignerMapper = np.vectorize(colorAssigner)
colors = colorAssignerMapper(y).tolist()
colors[0:10]
```

```python
plt.scatter(pcaX[:,0],pcaX[:,1],color=colors)
plt.show()
```
